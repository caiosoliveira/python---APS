```python
# Célula 1 — imports
import sys
sys.path.insert(0, '..')
from stt import transcribe_audio
from tts import speak
from response_engine import ResponseEngine
import IPython.display as ipd
```

```python
# Célula 2 — teste STT isolado com ficheiro de áudio
# Grava primeiro no 02_audio_input.ipynb e copia o caminho,
# ou usa o record_audio directamente aqui
from audio_input import record_audio

audio_path = record_audio(duration_seconds=4)
ipd.display(ipd.Audio(audio_path))
```

```python
# Célula 3 — transcrever o áudio gravado (gasta créditos STT)
transcricao = transcribe_audio(audio_path)
print(f"Transcrição: {transcricao}")
```

```python
# Célula 4 — obter resposta (zero créditos)
engine = ResponseEngine(responses_path="../responses.json")
resposta = engine.get_response(transcricao)
print(f"Resposta: {resposta}")
```

```python
# Célula 5 — sintetizar resposta em voz (gasta créditos TTS)
output = speak(resposta, output_path="../samples/resposta_teste.mp3")
ipd.display(ipd.Audio(output))
```
