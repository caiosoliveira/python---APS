```python
from dotenv import load_dotenv
import os
from elevenlabs.client import ElevenLabs

load_dotenv()
client = ElevenLabs(api_key=os.getenv("ELEVENLABS_API_KEY"))

# Verificar subscrição e créditos (SDK actualizado)
subscription = client.user.get()
print(f"Plano: {subscription.subscription.tier}")
print(f"Créditos usados: {subscription.subscription.character_count}")
print(f"Limite: {subscription.subscription.character_limit}")
```

```python
# Listar vozes disponíveis
voices = client.voices.get_all()
for v in voices.voices:
    print(f"Nome: {v.name:20} | ID: {v.voice_id} | Categoria: {v.category}")
```

```python
import IPython.display as ipd
import tempfile

def testar_voz(voice_id: str, nome: str, texto: str = "Olá, bem-vindo ao nosso sistema de atendimento automático."):
    print(f"A testar voz: {nome} ({len(texto)} caracteres)...")
    audio = client.text_to_speech.convert(
        voice_id=voice_id,
        text=texto,
        model_id="eleven_multilingual_v2",
    )
    tmp = tempfile.NamedTemporaryFile(suffix=".mp3", delete=False)
    for chunk in audio:
        tmp.write(chunk)
    tmp.flush()
    print(f"Voz {nome} gerada. Ouve abaixo:")
    return ipd.Audio(tmp.name, autoplay=False)

# Testa uma de cada vez para não desperdiçar créditos

# Testa só uma de cada vez
testar_voz("onwK4e9ZLuTAKqWW03F9", "Daniel")  # tom neutro, bom em PT

# Constantes do projecto - guardar aqui para referência
VOICE_ID_DANIEL = "onwK4e9ZLuTAKqWW03F9"
VOICE_ID_CAIO = "O3cxOovS9fnjvJc9LB1O"  # voz clonada do colega (referência)

print("Voice IDs guardados. Usa VOICE_ID_DANIEL nos módulos por enquanto.")
```
